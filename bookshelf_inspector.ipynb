```python
import pandas as pd
import numpy as np
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
users = pd.read_csv('gdrive/My Drive/BX-Users.csv', sep = "\t", encoding='latin')

display(users)
```

```python
# Rename the column with special characters
users.rename(columns={'User-ID;"Location";"Age"': 'User;Location;Age'}, inplace=True)

users['User;Location;Age'] = users['User;Location;Age'].str.replace('"', '')

# Splitting the column into separate columns
split_data = users['User;Location;Age'].str.split(';', expand=True)

# Assigning the split values to new columns
users['user_id'] = split_data[0]
users['location'] = split_data[1].str.title()
users['age'] = split_data[2]
users.drop('User;Location;Age', axis=1, inplace=True)

display(users)
```

```python
split_location = users['location'].str.split(',', expand=True)

users['city'] = split_location[0]
users['state'] = split_location[1]
users['country'] = split_location[2].fillna('None')
users.drop('location', axis=1, inplace=True)

users = users.drop_duplicates()
users = users.dropna()

display(users)
```

```python
books = pd.read_csv('gdrive/My Drive/BX-Books.csv', sep = "\t", encoding='latin')

display(books)
```

```python
books.rename(columns = {'ISBN;"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"'
                        : 'ISBN;Book-Title;Book-Author;Year-Of-Publication;Publisher;Image-URL-S;Image-URL-M;Image-URL-L'}, 
                        inplace=True) 
# books[['isbn', 'book_title', 'book_author', 'year-of-publication', 'publisher', 'image_url_s', 'image_url_m', 'image_url_l']]

books['ISBN;Book-Title;Book-Author;Year-Of-Publication;Publisher;Image-URL-S;Image-URL-M;Image-URL-L'] = books['ISBN;Book-Title;Book-Author;Year-Of-Publication;Publisher;Image-URL-S;Image-URL-M;Image-URL-L'].str.replace('"', '')

split_b_data = books['ISBN;Book-Title;Book-Author;Year-Of-Publication;Publisher;Image-URL-S;Image-URL-M;Image-URL-L'].str.split(';', expand=True)
books['isbn'] = split_b_data[0]
books['book_title'] = split_b_data[1]
books['book_author'] = split_b_data[2]
books['year-of-publication'] = split_b_data[3]
books['publisher'] = split_b_data[4]
books['image_url_s'] = split_b_data[5]
books['image_url_m'] = split_b_data[6]
books['image_url_l'] = split_b_data[7]
books.drop('ISBN;Book-Title;Book-Author;Year-Of-Publication;Publisher;Image-URL-S;Image-URL-M;Image-URL-L', axis=1, inplace=True)

books = books.drop_duplicates()
books = books.dropna()

display(books)
```

```python
ratings = pd.read_csv('gdrive/My Drive/BX-Book-Ratings.csv', sep = "\t", encoding='latin')

display(ratings)
```

```python
ratings.rename(columns = {'User-ID;"ISBN";"Book-Rating"' : 'user-id;isbn;book_rating'}, inplace=True) 

ratings['user-id;isbn;book_rating'] = ratings['user-id;isbn;book_rating'].str.replace('"', '')
ratings[['user_id', 'isbn', 'book_rating']] = ratings['user-id;isbn;book_rating'].str.split(";", expand=True)
ratings.drop('user-id;isbn;book_rating', axis=1, inplace=True)
ratings['book_rating'] = ratings['book_rating'].astype(int)

ratings = ratings.drop_duplicates()
ratings = ratings.dropna()

display(ratings)
```

```python
db = pd.merge(books, ratings, on='isbn')
db = pd.merge(users, db, on='user_id')

db = db.drop_duplicates()
db.replace(['None', 'NULL'], np.nan, inplace=True) 
db.replace(r'\\N/A\\', np.nan, regex=True, inplace=True) 
db.replace(r'^ $', np.nan, regex=True, inplace=True)
db.replace(r'^\s*$', np.nan, regex=True, inplace=True)
db = db.dropna()

display(db)
```

```python
# Check the max book rting
print(f'Max book rating: {db.book_rating.max()}')
```

```python
# Ghet top 4 books for each country with the max rating review
max_r_by_country = db.groupby(['country', 'isbn', 'book_title', 'image_url_l', 'book_author'])['book_rating'].max().reset_index(name='max_review')
max_r_by_country = max_r_by_country.groupby('country').apply(lambda x: x.nlargest(10, 'max_review')).reset_index(drop=True)

display(max_r_by_country)
```

```python
def get_book_description(isbn):
  url = f'https://www.googleapis.com/books/v1/volumes?q=isbn:{isbn}'
  response = requests.get(url)
  data = response.json()

  if 'items' in data and len(data['items']) > 0:
      book_info = data['items'][0]['volumeInfo']
      description = book_info.get('description', 'No description available')
      return description
  else:
      return 'Book not found'


# List of ISBNs
list_of_ISBN = max_r_by_country['isbn'].values.tolist()

# Create an empty DataFrame
book_description = pd.DataFrame(columns=['ISBN', 'Description'])

# Iterate over the ISBNs and fetch the descriptions
for isbn in list_of_ISBN:
    description = get_book_description(isbn)
    book_description = book_description.append({'ISBN': isbn, 'Description': description}, ignore_index=True)

display(book_description)
```
